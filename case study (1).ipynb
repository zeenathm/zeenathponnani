1.Reading the dataset to python environment

```python

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics 
sns.set()
filename='cars_data.csv' 
df=pd.read_csv(filename)
df

```

2.Checking for the null values present in the dataset

```python
df.info()
```

3.Bar graph of male vs female buyers participated in the sales

```python
df['Buyer Gender'].value_counts()
```

```python
sns.set(font_scale=1.4)
df['Buyer Gender'].value_counts().plot(kind='bar',figsize=(7,6),rot=0,color=['red','blue'])
plt.xlabel("Gender",labelpad=14)
plt.ylabel("count of people",labelpad=14)
plt.title('Bar graph of male vs female buyers participated in the sales')
```

In the above bar graph we can understand that the male and female participation in the sale

4.The top  5 cars based on their sales price

```python
top_sale_price=df.sort_values(by=['Sale Price']).tail()
top_cars_df=[top_sale_price["Make"]+" "+top_sale_price["Model"],top_sale_price['Sale Price']]
headers=["Car Name","Sale Price"]
top_cars=pd.concat(top_cars_df,axis=1,keys=headers)



```

```python
top_cars
```

The above are based on their sales price

5.The least 5 cars based on their resell price

```python
least_resell_price=df.sort_values(by=['Resell Price']).head()
least_cars_df=[least_resell_price["Make"]+" "+least_resell_price["Model"],least_resell_price['Resell Price']]
headers=["Car name","Resell Price"]
least_cars=pd.concat(least_cars_df,axis=1,keys=headers)
```
